```python
# ✅ Import libraries
from gtts import gTTS
import os
from IPython.display import Audio
```

```python
# ✅ Use the final_description from previous step
final_description = "A person sitting in a wheelchair at a bus stop. The text on the image reads: Welcome to VisionVoice! This is a test image with some printed text."
```

```python
# ✅ Convert to speech
tts = gTTS(text=final_description, lang='en')
audio_path = "../outputs/visionvoice_audio.mp3"
tts.save(audio_path)
```

```python
# ✅ Play audio
Audio(audio_path)
```
